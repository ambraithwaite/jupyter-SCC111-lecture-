```python
#get april from each year
april_rainfall = df.loc[(df.index >= '04-01') & (df.index <= '04-30')].copy()

#assign a 7-day trial
april_rainfall['Week_label'] = pd.cut(
    april_rainfall.index.str[3:].astype(int),
    bins=[0,7,14,21,28,30],
    labels=['Apr 1-7', 'Apr 8-14','Apr 15-21','Apr 22-28','Apr 29-30']
)

```
